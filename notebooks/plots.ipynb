```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
spotify_df = pd.read_csv("../data/Cleaned_Spotify_2024_Global_Streaming_Data.csv")
spotify_df.head()
```

```python
avg_streams = spotify_df.groupby("Genre")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_streams
```

```python
plt.bar(avg_streams.index, avg_streams.values)
plt.show()
```

```python
plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")

plt.show()
```

```python
plt.figure(figsize=(10,6)) # x y size inches
plt.bar(avg_streams.index, avg_streams.values)

ax = plt.gca()
ax.spines['top'].set_visible(False)     # black line top of graph
ax.spines['right'].set_visible(False)   # black line right of graph

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45, ha="right")    # ha = horizontal alignment

plt.tight_layout()
plt.show()
```

```python
avg_artist_streams = spotify_df.groupby("Artist")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_artist_streams
```

```python
plt.figure(figsize=(10,6))

colors = ["#FED321"]*len(avg_artist_streams)

max_index = avg_artist_streams.idxmax()
highlight_position = avg_artist_streams.index.get_loc(max_index)

colors[highlight_position] = "#321def"

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width = 0.8 )

ax = plt.gca()
ax.spines['top'].set_visible(False)     # black line top of graph
ax.spines['right'].set_visible(False)   # black line right of graph

plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Artist")
plt.xticks(rotation=45, ha="right")    # ha = horizontal alignment

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))

special_artist = "Ariana Grande"
colors = ["#FED321" if artist == special_artist else "gray" for artist in avg_artist_streams.index]

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width = 0.8 )

ax = plt.gca()
ax.spines['top'].set_visible(False)     # black line top of graph
ax.spines['right'].set_visible(False)   # black line right of graph

plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Ariana Grande Ranks 2nd in Total Average Streams")
plt.xticks(rotation=45, ha="right")    # ha = horizontal alignment

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

featured_genre = "Pop"

colors = ["#08cc15" if genre == featured_genre else "#d1d5d1" for genre in spotify_df["Genre"]]

plt.scatter(
  spotify_df["Streams Last 30 Days (Millions)"], 
  spotify_df["Total Streams (Millions)"],
  color = colors,
  alpha = 0.8,
  s = 25)

plt.xlabel("Streams this Month")
plt.ylabel("Total Streams")

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.set_title("Do Pop Songs Achieve Higher Monthly Listens or Total Streams?", fontsize = 18, pad = 35)

for spine in ax.spines.values():
  spine.set_linewidth(.25)
  spine.set_alpha(0.5)

plt.tight_layout()
plt.show()
```

```python
category = "Platform Type"
values = "Total Streams (Millions)"

platform_totals = spotify_df.groupby(category)[values].sum()
platform_totals
```

```python
plt.figure(figsize=(10, 6))
plt.pie(platform_totals)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.pie(
  platform_totals,
  labels = platform_totals.index,
  startangle=110, 
  colors= ["gray", "blue"])

plt.title("Near Even Split Between Premium and Free Users")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.pie(
  platform_totals,
  labels = platform_totals.index,
  startangle=110, 
  colors= ["gray", "blue"],
  autopct="%1.1f%%")

plt.legend(
  labels = platform_totals.index,
  loc="lower right",
  title = "Platform")

plt.title("Near Even Split Between Premium and Free Users")
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))

wedges, texts, autotexts = ax.pie(
  platform_totals,
  autopct="%1.1f%%",
  startangle=110, 
  colors= ["gray", "blue"],)

ax.set_title(
  "Near Even Split Between Premium and Free Users",
  fontsize=18,
  pad=20,
  loc="center")

ax.legend(
  wedges,
  platform_totals.index,
  loc = "lower center",
  bbox_to_anchor = (0.5, 0.9),
  ncol = 2)

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))

wedges, texts, autotexts = ax.pie(
  platform_totals,
  autopct="%1.1f%%",
  startangle=110, 
  colors= ["gray", "blue"],)

ax.set_title(
  "Near Even Split Between Premium and Free Users",
  fontsize=25,
  pad=15,
  loc="center")

for autotext in autotexts: 
  autotext.set_fontsize(16)
  autotext.set_color("white")
  autotext.set_weight("bold")
  
ax.legend(
  wedges,
  platform_totals.index,
  loc = "lower center",
  bbox_to_anchor = (0.5, 0.9),
  ncol = 2)

plt.tight_layout()
plt.show()
```
